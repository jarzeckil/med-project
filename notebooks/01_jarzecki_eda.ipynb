## Correlation

```python
corr = diabetes.corr(numeric_only=True)
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(25, 25))
sns.heatmap(
    corr,
    annot=True,
    vmin=-1,
    vmax=1,
    mask=mask,
    linewidths=0.5,
)
```
