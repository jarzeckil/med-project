Zakodowanie zmiennych porzÄ…dkowych

```python
education_categories = ['No formal', 'Highschool', 'Graduate', 'Postgraduate']
income_categories = ['Low', 'Lower-Middle', 'Middle', 'Upper-Middle', 'High']
smoking_categories = ['Never', 'Former', 'Current']
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[education_categories, income_categories, smoking_categories])

diabetes[['education_level_encoded', 'income_level_encoded', 'smoking_status_encoded']] = encoder.fit_transform(
    diabetes[['education_level', 'income_level', 'smoking_status']]
)
```

```python
diabetes[['education_level', 'education_level_encoded', 'income_level', 'income_level_encoded', 'smoking_status', 'smoking_status_encoded']].head(10)
```

```python
diabetes = diabetes.drop(['education_level', 'income_level', 'smoking_status'], axis=1)
```

```python
diabetes = pd.get_dummies(data=diabetes, columns=['employment_status', 'gender', 'ethnicity'], drop_first=True)
```

Correlation

```python
corr = diabetes.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))


plt.figure(figsize=(25, 25))
sns.heatmap(diabetes.corr(), annot=True, vmin=-1, vmax=1, mask=mask, linewidths=.5,)
```
